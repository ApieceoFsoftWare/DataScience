## Çok Değişkenli Aykırı Gözlem Analizi

#### Local Outlier Factor

Gözlemleri bulundukları konumda yoğunluk tabanlı skorlayarak buna göre aykırı değer olabilecek değerleri tanımlayabilmemize imkan sağlıyor.

Bir noktanın local yoğunluğu bu noktanın komşuları ile karşılaştırılıyor. Eğer bir nokta komşularının yoğunluğundan anlamlı şekilde düşük ise bu nokta komşularından daha seyrek bir bölgede bulunuyordur yorumu yapılabiliyor. Dolayısıyla burada bir komşuluk yapısı söz konusu. Bir değerin çevresi yoğun değilse demek ki bu aykırı değerdir şeklinde değerlendiriliyor.

```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds = diamonds.select_dtypes(include =["float64","int64"])
df= diamonds.copy()
df=df.dropna()
df.head()
```

```python
import numpy as np
from sklearn.neighbors import LocalOutlierFactor
```

```python
clf = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
```

```python
clf.fit_predict(df)
```

```python
df_scores = clf.negative_outlier_factor_
```

```python
df_scores[0:10]
```

```python
np.sort(df_scores)[0:20]
```

```python
np.sort(df_scores)[13]
```

```python
esik_deger = np.sort(df_scores)[13]
```

```python
aykiri_tf = df_scores > esik_deger
```

```python
aykiri_tf
```

#### Silme Yöntemi

```python
yeni_df = df[df_scores > esik_deger]
yeni_df
```

```python
df[df_scores < esik_deger]
```

#### Baskılama Yöntemi

```python
df[df_scores == esik_deger]
```

```python
baski_degeri = df[df_scores == esik_deger] 
```

```python
aykiri_deger = df[~aykiri_tf]
```

```python
aykiri_deger.to_records(index=False)
```

```python
res = aykiri_deger.to_records(index=False)
```

```python
res[:] = baski_degeri.to_records(index=False)
```

```python
res
```

```python
df[~aykiri_tf]
```

```python
import pandas as pd
df[~aykiri_tf] = pd.DataFrame(res, index=df[~aykiri_tf].index)
```
