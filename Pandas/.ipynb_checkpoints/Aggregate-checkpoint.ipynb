**Aggregate**

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
                    'gruplar'  :["A","B","C","A","B","C"],
                    'degisken1':[10,23,33,22,11,79],
                    'degisken2':[100,312,546,422,789,811]},
                 
                    columns =["gruplar","degisken1","degisken2"])
df
```

**Aggregate**

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min",np.median, max]) # burada min ve max'ı tırnak içinde yazmamıza gerek yok fakat np.median ı tırnak içine alamayız
                                                        #Bunun nedeni median metodunu o şekilde bulamamasından kaynaklıdır.
```

```python
df.groupby("gruplar").aggregate({"degisken1": "min",
                                 "degisken2": "max"})
```
