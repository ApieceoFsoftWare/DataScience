## Hiyerarik Kmeleme Uygulama 1

```python
from warnings import filterwarnings
filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("USArrests.csv").copy()
df.index = df.iloc[:,0]
df =df.iloc[:, 1:5]
df = df.rename_axis(None, axis=0)
df.head()
```

```python
from scipy.cluster.hierarchy import linkage
hc_complete = linkage(df, "complete")
hc_average = linkage(df, "average")
hc_single = linkage(df, "single")
```

```python
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(15,10))
plt.title('Hiyerarik Kmeleme - Dendrogram')
plt.xlabel('ndexler')
plt.ylabel('Uzaklk')

dendrogram(hc_complete, leaf_font_size=10);
```

```python
from scipy.cluster.hierarchy import dendrogram

plt.figure(figsize=(15,10))
plt.title('Hiyerarik Kmeleme - Dendrogram')
plt.xlabel('ndexler')
plt.ylabel('Uzaklk')

dendrogram(hc_complete, truncate_mode="lastp", p=4, show_contracted=True);
```

### Optimum Kme Says

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 4, affinity="euclidean", linkage="ward")

cluster.fit_predict(df)
```

```python
pd.DataFrame({"Eyaletler": df.index, "Kumeler": cluster.fit_predict(df)})[0:10]
```
