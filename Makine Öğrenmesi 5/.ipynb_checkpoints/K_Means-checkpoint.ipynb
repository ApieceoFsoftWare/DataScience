```python
from warnings import filterwarnings
filterwarnings('ignore')

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sp
from sklearn.cluster import KMeans
```

## K Means

```python
df = pd.read_csv('USArrests.csv').copy()
df.head()
```

```python
df.index = df.iloc[:,0]
```

```python
df.index
```

```python
df.head()
```

```python
df = df.iloc[:,1:5]
df.head()
```

```python
df = df.rename_axis(None, axis=0)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

## K-Means Model ve Grselletirme

```python
kmeans = KMeans(n_clusters = 4)
kmeans.get_params()
```

```python
k_fit = kmeans.fit(df)
```

```python
k_fit.cluster_centers_
```

```python
k_fit.labels_
```

```python
#grselletirme
```

```python
kmeans = KMeans(n_clusters=2)
k_fit = kmeans.fit(df)
```

```python
kumeler = k_fit.labels_
```

```python
plt.scatter(df.iloc[:,0], df.iloc[:,1], c=kumeler, s=50, cmap="viridis")

merkezler = k_fit.cluster_centers_
plt.scatter(merkezler[:,0], merkezler[:,1], c="black", s=200, alpha=0.5);
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
kmeans = KMeans(n_clusters=3)
k_fit = kmeans.fit(df)
kumeler = k_fit.labels_
merkezler = kmeans.cluster_centers_
```

```python
plt.rcParams['figure.figsize'] = (16,9)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2]);
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(df.iloc[:,0],df.iloc[:,1],df.iloc[:,2], c=kumeler)
ax.scatter(merkezler[:,0],merkezler[:,1],merkezler[:,2], marker="*", c="#050505", s=1000);
```

```python
#kmeler ve gzlem deerleri

kmeans = KMeans(n_clusters=3)
k_fit = kmeans.fit(df)
kumeler = k_fit.labels_
```

```python
pd.DataFrame({"Eyaletler": df.index, "Kmeler" : kumeler})[0:10]
```

```python
df["kume_no"] = kumeler
df.head()
```

```python
df["kume_no"] = df["kume_no"] +1
```
