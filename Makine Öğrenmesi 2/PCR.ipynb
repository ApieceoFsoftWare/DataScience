## PCR MODEL

```python
import pandas as pd
hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
df.head()   
```

```python
df.info()
```

```python
df.describe().T
```

```python
dms = pd.get_dummies(df[['League','Division','NewLeague']])
```

```python
dms.head()
```

```python
y = df["Salary"]
```

```python
X = df.drop(["Salary","League","Division","NewLeague"], axis=1).astype("float64")
```

```python
X.head()
```

```python
X1 = pd.concat([X, dms[["League_N","Division_W","NewLeague_N"]]] , axis = 1)
X1.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
```

```python
X1_Train, X1_Test, Y_Train, Y_Test = train_test_split(X1,
                                                      y,
                                                      test_size=0.25,
                                                      random_state=42)

print("X1_Train", X1_Train.shape)

print("X1_Test", X1_Test.shape)

print("Y_Train", Y_Train.shape)

print("Y_Test", Y_Test.shape)

training = df.copy()

print("Training", training.shape)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
pca = PCA()
```

```python
X1_reduce_train = pca.fit_transform(scale(X1_Train))
```

```python
X1_reduce_train[0:1,:]
```

```python
import numpy as np
np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)[0:10]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
pcr_model = lm.fit(X1_reduce_train, Y_Train)
```

```python
pcr_model.intercept_
```

```python
pcr_model.coef_
```

## PCR TAHMƒ∞N

```python
y_pred = pcr_model.predict(X1_reduce_train)
y_pred[0:5]
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.sqrt(mean_squared_error(Y_Train,y_pred))
```

```python
df["Salary"].mean()
```

```python
r2_score(Y_Train,y_pred)
```

```python
pca2 = PCA()
```

```python
X1_reduced_test = pca2.fit_transform(scale(X1_Test))
```

```python
y_pred = pcr_model.predict(X1_reduced_test)
```

```python
np.sqrt(mean_squared_error(Y_Test,y_pred)) # Test Hatasƒ±
```

## PCR MODEL TUNING

```python
lm = LinearRegression()
pcr_model = lm.fit(X1_reduce_train[:,:], Y_Train)
y_pred = pcr_model.predict(X1_reduced_test[:,:])

print(np.sqrt(mean_squared_error(Y_Test, y_pred)))
```

```python
from sklearn import model_selection
```

```python
cv_10 = model_selection.KFold(n_splits=10,
                              shuffle=True,
                              random_state=1)
```

```python
lm = LinearRegression()
```

```python
RMSE = []
```

**A≈üaƒüƒ±daki kod : Herbir bile≈üen sayƒ±sƒ± i√ßin gidip model kuracak, herbir bile≈üen i√ßin K Katlƒ± CrossValidation kurarak buradaki hata deƒüerlerini elde edip daha sonra bize hangi bile≈üen sayƒ±sƒ±nƒ±n daha az hata olu≈üturduƒüu bilgisini sunacak.**

```python
for i in np.arange(1, X1_reduce_train.shape[1] + 1):
    
    score = np.sqrt(-1*model_selection.cross_val_score(lm, 
                                                       X1_reduce_train[:,:i], 
                                                       Y_Train.ravel(), 
                                                       cv=cv_10, 
                                                       scoring='neg_mean_squared_error').mean())
    RMSE.append(score)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(RMSE, '-v')
plt.xlabel('Bile≈üen Sayƒ±sƒ±')
plt.ylabel('RMSE')
plt.title('Maa≈ü Tahmin Modeli ƒ∞√ßin PCR Model Tuning');
```
