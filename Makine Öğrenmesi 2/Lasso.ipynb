## Lasso Regresyon

Amaç; hata kareler toplamının minimize eden katsayıları bu kat sayılara bir ceza uygulayarak bulmaktır. Ridge Regresyonun aynısıdır.
Farkı ise ceza uygulama işlemini abartarak katsayılarının cezalarını onları sıfır yapacak şekilde uygulamaktadır.

**Ridge Regresyonunun ilgili-ilgisiz tüm değişkenleri modelde bırakma dezavantajını gidermek için önerilmiştir.**

**Lasso'da katsayıları sıfıra yaklaştırır.**

**Fakat L1 normu λ yeteri kadar büyük olduğunda bazı kaysayıları sıfır yapar. Böylece değişken seçimi yapmış olur.**

**λ doğru şeçilmesi çok önemlidir, burada da CV kullanılır.**

**Ridge ve Lasso yöntemleri birbirinden üstün değildir.**

### λ Parametresinin Belirlenmesi

λ'nın sıfır olduğu yer EKK'dır. HKT'yi mininmum yapan λ'yı arıyoruz.

λ için belirli değerleri içeren kümeler seçilir ve her birisi için Cross Validation test hatası hesaplanır.

En küçük Cross Validation'ı veren λ ayar parametresi olarak seçilir.

Son olarak seçilen bu λ ile model yeniden tüm gözlemlere fit edilir.

## Lasso Regresyon

#### Model

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict

hit = pd.read_csv("Hitters.csv")
df = hit.copy()
df = df.dropna()
dms = pd.get_dummies(df[["League","Division","NewLeague"]])
y = df["Salary"]
X_ = df.drop(["Salary","League","Division","NewLeague"],axis=1).astype('float64')
X = pd.concat([X_,dms[["League_N","Division_W","NewLeague_N"]]], axis=1)
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, y , test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha=0.1).fit(X_Train, Y_Train)
```

```python
lasso_model
```

```python
lasso_model.coef_
```

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
lasso = Lasso()
lambdalar = 10**np.linspace(10,-2,100)*0.5
katsayilar = []
for i in lambdalar: 
    lasso.set_params(alpha=i)
    lasso.fit(X_Train, Y_Train)
    katsayilar.append(lasso.coef_)
    
ax = plt.gca()
ax.plot(lambdalar*2, katsayilar)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel("alpha")
plt.ylabel("weights")
```

### Tahmin

```python
lasso_model.predict(X_Test)
```

```python
y_pred = lasso_model.predict(X_Test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
np.sqrt(mean_squared_error(Y_Test, y_pred))
```

### Model Tuning

```python
from sklearn.linear_model import LassoCV
```

```python
lasso_cv_model = LassoCV(alphas=None, cv=10, max_iter=10000, normalize=True).fit(X_Train,Y_Train)
```

```python
lasso_cv_model.alpha_
```

```python
lasso_tuned = Lasso(alpha=lasso_cv_model.alpha_)
```

```python
lasso_tuned.fit(X_Train, Y_Train)
```

```python
y_pred = lasso_tuned.predict(X_Test)
```

```python
np.sqrt(mean_squared_error(Y_Test, y_pred))
```
